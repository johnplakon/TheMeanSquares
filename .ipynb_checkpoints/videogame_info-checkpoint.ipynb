```python
#Dependencies
import urllib.request
import requests
import pandas as pd
import numpy as np
# Import BS library so we can parse HTML and XML document
from bs4 import BeautifulSoup as bs
```

```python
# specify which url/web pages we are scraping
url_best_selling = "https://en.wikipedia.org/wiki/List_of_best-selling_video_games"
url_high_gross = "https://vgsales.fandom.com/wiki/List_of_highest-grossing_video_games"
```

```python
# open the  best_selling url using urllib.request and put the html into variable
page = urllib.request.urlopen(url_best_selling)
```

```python
# parse the HTML from our url into the BS parse tree format
best_sell_soup = bs(page, 'lxml')
```

```python
# print(best_sell_soup.prettify())
```

```python
# use 'find_all' function to bring back all instance of the table tag
best_sell_tables=best_sell_soup.find_all('table')
# best_sell_tables
```

```python
sales_table=best_sell_soup.find('table', class_='wikitable plainrowheaders sortable')
#sales_table
```

```python
# To look thru the rows, setup array to store strings
A = []
B = []
C =[]
D = []
E =[]
F = []

for row in sales_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==8:
        t_a=cells[0].findAll('a')
        A.append(t_a[0].contents[0])
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
        D.append(cells[3].find(text=True))
        E.append(cells[4].find(text=True))
        F.append(cells[5].find(text=True))
        
print(F[:5])        
```

```python
df=pd.DataFrame(A, columns=['Title'])
df['Sales']=B
df['Platform']=C
df['Release_date']=D
df['Developer']=E
df['Publisher']=F
df
```
